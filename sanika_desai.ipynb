```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Mall_Customers.csv')
df.head()
print("Shape:", df.shape)

print("\nInfo:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

df.describe()

df.columns = ['CustomerID', 'Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']

sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()

sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Age Distribution')
plt.show()

sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=df, hue='Gender')
plt.title('Income vs Spending Score')
plt.show()

sns.boxplot(x='Gender', y='Annual Income (k$)', data=df)
plt.title('Income Distribution by Gender')
plt.show()


```

### Key Insights from EDA:

- Female customers are slightly more than male customers.
- Majority of customers are in the age group of 25 to 40.
- Income and spending score don't have a strong correlation.
- Both high-income and low-income customers have varied spending patterns.
- Males tend to have a slightly higher average income than females.
